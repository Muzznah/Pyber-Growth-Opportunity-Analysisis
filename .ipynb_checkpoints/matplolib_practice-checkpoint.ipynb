```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
# st x-axis to a list of strings for each month.
x_axis=["Jan","Feb","Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#chain plt with plot function and give values for graph.
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#create the plot and add a label for the legend.
plt.plot(x_axis,y_axis,marker="*", color="purple", linewidth=2, label="Boston")
#create labels for x-axis and y axis.
plt.xlabel("Date")
plt.ylabel("Fare($)")
#set the y limit betwenn 0-45.
plt.ylim(0,45)
#create a title.
plt.title("PyBer Fare by Month")
#add grid
plt.grid()
#add the legend
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis,color="green",marker="D",linewidth=2, label="Boston")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_ylim(0,45)
ax.grid()
ax.legend()
```

```python
#create the bar plot.
plt.bar(x_axis,y_axis, color="green", label="Boston")
#create the labels for axis
plt.xlabel("Date")
plt.ylabel("Fare($)")
#create title
plt.title("Pyber Fare by Month")
plt.legend()
```

```python
#create the horizonta bar chart
plt.barh(x_axis,y_axis,color="magenta", label="Boston")
#invert y axis
plt.gca().invert_yaxis()
#add axis
plt.xlabel("Fare($)")
plt.ylabel("Date")
#add title
plt.title("PyBar Fare by Month")
#set y axis limits
#add legend
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with ax.plt()
fig, ax =plt.subplots()
ax.bar(x_axis,y_axis, color="gold", label="New York")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.set_ylim(0,45)
ax.legend()
```

```python
# create a horizontal plot.
fig,ax=plt.subplots()
ax.barh(x_axis,y_axis, color="cyan", label="Chicago")
#invert the y axis.
plt.gca().invert_yaxis()
#add axis.
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
#add title.
ax.set_title("PyBer Fare by Month")
#add legend.
ax.legend()
```

```python
#plot a scatter plot
plt.plot(x_axis,y_axis,"o")
```

```python
#plot a scatter plot
plt.scatter(y_axis,x_axis, c="r", label="Chicago")
# Add a legend for the city of Chicago.
plt.legend()
# Add a title and axes labels.
plt.title("PyBer Fares by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Switch the axis so that the Fare($) data is on the x-axis.
# Add a limit for the x-axis data.
plt.xlim(0,45)
# Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()
```
