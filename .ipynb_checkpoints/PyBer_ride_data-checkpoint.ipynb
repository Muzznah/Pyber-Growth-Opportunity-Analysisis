```python
#import dependencies.
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#load in csv
pyber_ride_df= pd.read_csv("Resources\PyBer_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# adjust the x-ticks to show all the months.
x_axis=np.arange(len(pyber_ride_df))
tick_location=[value for value in x_axis]
#plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_location,pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind="bar")
plt.show()
```

```python
# Create a bar chart.
#Import dependencies.
import statistics

#state axis.
x_axis=pyber_ride_df["Month"]
y_axis=pyber_ride_df["Avg. Fare ($USD)"]

# Add  stdev for error bars.
stdev=statistics.stdev(y_axis)
stdev
```

```python
#Plot bar chart with error bar and caps
pyber_ride_df.plot.bar(x="Month",y="Avg. Fare ($USD)", yerr=stdev, capsize=3, color='skyblue')

# Set the y-axis increment to every $5.
plt.yticks=np.arange(0,51,step=5.0)

# Rotate the labels on the x-axis to horizontal.
plt.xticks(rotation=0)

plt.show()
```
