```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with ax.plt()
fig, ax =plt.subplots()
ax.bar(x_axis,y_axis, color="gold", label="New York")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.set_ylim(0,45)
ax.legend()
```

```python
# create a horizontal plot.
fig,ax=plt.subplots()
ax.barh(x_axis,y_axis, color="cyan", label="Chicago")
#invert the y axis.
plt.gca().invert_yaxis()
#add axis.
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
#add title.
ax.set_title("PyBer Fare by Month")
#add legend.
ax.legend()
```

```python
#plot a scatter plot
plt.plot(x_axis,y_axis,"o")
```

```python
#plot a scatter plot
plt.scatter(y_axis,x_axis, c="r", label="Chicago")
# Add a legend for the city of Chicago.
plt.legend()
# Add a title and axes labels.
plt.title("PyBer Fares by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Switch the axis so that the Fare($) data is on the x-axis.
# Add a limit for the x-axis data.
plt.xlim(0,45)
# Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()
```

```python
#modifying the y axis values to make the bubbles bigger
large_y_axis=[]
for num in y_axis:
    large_y_axis.append(num*3)

large_y_axis
```

```python
#make a scatter plot using new data.
plt.scatter(x_axis,y_axis, s=large_y_axis)
```

```python
#usling list comprehension instead of for loop.
plt.scatter(x_axis,y_axis, s=[i*3 for i in y_axis])
```

```python
# make a scatter plot with object oriented.
fig,ax=plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
#create a bubble chart using object-oriented interface.
# Change the color of the markers to sky blue.
# Change the size of the markers to 5 times each data point.
# Make the color 20% transparent.
# Add a black edge color to the circles.
# Make the linewidth of the circles 2 points.
# Add a legend for the city of Boston.
fig,ax=plt.subplots()
ax.scatter(y_axis,x_axis, color="skyblue",edgecolor="black",linewidth=2,alpha=.80,s=[i*5 for i in y_axis], label="Boston")
ax.legend()
# Add a title.
ax.set_title("PyBer Fares by Month")
# Switch the axis so that the Fare($) data is on the x-axis.
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
# Add a limit for the x-axis data.
ax.set_xlim(0,45)
# Invert the y-axis so that January is at the top
plt.gca().invert_yaxis()
```

```python
#Create Piecharts.
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
#Set colors for the wedges.
colors=["pink","plum",'mediumpurple','slateblue','skyblue','grey','greenyellow','yellow','coral','orange','gold','orangered']
#increase the size of pie.
plt.subplots(figsize=(8,8))
#add percentage and explode july.
explode_values=(0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis, labels=x_axis,autopct="%.1f%%", explode=explode_values,colors=colors)
plt.show()
```

```python
# Create a Pie Chart Using the Object-Oriented Interface.
# Explode the two highest percentage months.
explode=[0,0,0.2,0,0,0,0.2,0,0,0,0,0]
# Add new colors of your own choosing or use the colors from the previous pie chart.
colors=["pink","plum",'mediumpurple','slateblue','skyblue','grey','greenyellow','yellow','coral','orange','gold','orangered']
# Add a percentage to one decimal place to each wedge of the pie.
# Increase the figure size to 8x8.(can also do by setting r=2)
# Add a shadow.
# Add a start angle so that January is at the top.
# Reverse the order so that the month order is in a clockwise direction.
fig,ax=plt.subplots(figsize=(8,8))
ax.pie(y_axis,
      labels=x_axis,
      autopct="%.1f%%",
      explode=explode,
      shadow=True,
      startangle=90,
      colors=colors,
      counterclock=False)
plt.show()
```
